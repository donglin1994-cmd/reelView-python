```python
#load in python libraries to access data and build visualization
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import csv file and get the first five row of dataset
reelView = pd.read_csv(r"C:\Users\DongqinL\Downloads\archive (1)\ReelView.csv")
reelView.head()
```

```python
# convert dataframe use best possible datatype
reelView=reelView.convert_dtypes()
```

```python
# check data type for all columns
reelView.dtypes
```

```python
# finding the number of occurance of a specified value in genre column 
reelView['genre'].value_counts()
```

```python
# drop all rows where at least one column misssing
reelView.dropna()
```

```python
data= reelView['stars_rating']
print(type(data))
```

```python
# loop through value in stars_rating column 
# create an empty list to hold values for new column name stars_rating_scale
list1=[]
for i in range(len(data)):
    if data.iloc[i]>3.5 and data.iloc[i]<=4:
        list1.append('Excellent')
    elif data.iloc[i]>3 and data.iloc[i]<=3.5:
        list1.append('Good')
    elif data.iloc[i]>2.5 and data[i]<=3:
        list1.append('Not Bad')
    elif data.iloc[i]<=2.5:
        list1.append('Awful')
    else:list1.append('NaN')
    
print(list1)
```

```python
#add a new column at position 2 
reelView.insert(2,'stars_rating_scale',list1)
print(reelView)
```

```python
# first group based on stars_rating
df=reelView.groupby(['stars_rating','title'])
df.first()
```

```python
# finding the 1st and 3rd quartile respectively 
q1 = np.quantile(data, 0.25)
q3 = np.quantile(data, 0.75)
med = np.median(data)
# finding the iqr region
iqr = q3-q1
 # finding upper and lower whiskers
upper_bound = q3+(1.5*iqr)
lower_bound = q1-(1.5*iqr)
print(iqr, upper_bound, lower_bound)
outliers = data[(data <= lower_bound) | (data >= upper_bound)]
print('The following are the outliers in the boxplot:{}'.format(outliers))
```

```python
# style and customized boxplot with seaborn 
ax=sns.boxplot(y=reelView['run_time'],x=reelView['stars_rating_scale'],width=0.3,linewidth=1.5)
plt.show()
```
